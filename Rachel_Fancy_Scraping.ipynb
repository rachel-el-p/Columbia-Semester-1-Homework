### Patterns for infinite scroll and pagination

````python
#<button class="btn btn-default next">
#So you're going to ask playwright to find with one of these classes

#page.locator(".btn")
#page.locator(".btn-default")
#page.locator(".next")

#BeautifulSoup looks for an exact match, not care about how CSS selectors work 
#doc.find_all(class_="btn btn-default next")
#Unless you're using .select
#doc.select(class_=".btn.btn-default next")

#Find me something with the class of btn-default AND next
await page.locator (".btn-default.next").click()
````

````python
#instead of saying "do this 322 time" for _ in range(322), you say "do this forever"

#Use a WHILE LOOP to keep clicking this button:
while True:
    try:
        print("I clicked it")
        await page.locator (".btn-default.next").click()
    except:
        print("I couldn't click it")
        #pass means "don't do anything"
        #break is going to exit the loop and not print an error 
        break
````

````python
# One dataframe for each page
dataframes = []

# instead of saying "do this 322 times"
# for _ in range(322)
# you say "do this forever"
while True:
    try:
        # Do your scraping in here
        tables = pd.read_html(await page.content())
        df = tables[0]
        
        # Add the dataframe to the list of dataframes
        dataframes.append(df)
        
        print("Trying to click the button")
        # Attempt to click the button
        # If it doesn't show up in 2 seconds, give up
        await page.locator(".btn-default.next").click(timeout=2000)
    except:
        print("Could not click the button")
        # Oh no, we got an error, no button found!
        # pass means "don't do anything"
        # 'break' is going to exit the loop
        # and not print an error
        break

print("Now we have", len(dataframes), "dataframes")
````

````python
# Combine all of the pages
df = pd.concat(dataframes, ignore_index=True)
df.shape
````
