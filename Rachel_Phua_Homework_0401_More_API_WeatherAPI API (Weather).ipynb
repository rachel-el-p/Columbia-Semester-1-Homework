## 8) For the next 24-ish hours in Miami, what percent of the time is the temperature above 85 degrees?

- *Tip: You might want to read up on [looping patterns](http://jonathansoma.com/lede/foundations-2017/classes/data%20structures/looping-patterns/)*

```python
response=requests.get("http://api.weatherapi.com/v1/forecast.json?key=3f4c9d67a228443e972203755220511&q=miami&days=1&aqi=no&alerts=no")
miami_weather=response.json()

total = 0
for temperature in miami_weather['forecast']['forecastday']:
    if (temperature['hour']['temp_f'])>85:
        total=total+temperature

print ((total/24)x100)
```
